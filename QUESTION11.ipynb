A)PROBABILITY OF A STUDENT BEING A GRADUATE UNDER VARIABLE TARGET.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/EHMAH/Documents/Students.csv", delimiter=';')
total_students = len(df)
#number of graduates
graduates = len(df[df['Target'] == 'Graduate'])

#probability
probability = graduates / total_students
print('Total number of students:', total_students)
print('Number of graduates:', graduates)
print('Probability of being a graduate: {:.4f}'.format(probability))


# Using piechart to visualize
target_counts = df['Target'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Student Outcomes')
plt.show()
```

B)PROBABILITY OF A STUDENT BEING A DROPOUT WITH A MARITAL STATUS OF 1

```python

#Filtering dataset for students with a marital status of 1
marital_status_1 = df[df['Marital status'] == 1]

#Total number of students with marital status 1
total_marital_status_1 = len(marital_status_1)

#Number of dropouts among these students
dropouts_marital_status_1 = len(marital_status_1[marital_status_1['Target'] == 'Dropout'])

#Probability
probability_dropout_marital_status_1 = dropouts_marital_status_1 / total_marital_status_1

print('Total number of students with marital status 1:', total_marital_status_1)
print('Number of dropouts with marital status 1:', dropouts_marital_status_1)
print('Probability of being a dropout with marital status 1: {:.4f}'.format(probability_dropout_marital_status_1))
print('Probability as percentage: {:.2f}%'.format(probability_dropout_marital_status_1 * 100))

```

C)THE RELATIONSHIP BETWEEN THE VARIABLES TARGET AND MOTHER QUALIFICATION.

```python

#cross-tabulation of Target and Mother's Qualification
cross_tab = pd.crosstab(df['Mother\'s qualification'], df['Target'], normalize='index')

#Relationship using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, annot=True, cmap='coolwarm', cbar_kws={'label': 'Proportion'})
plt.title('Relationship between Target and Mother\'s Qualification')
plt.xlabel('Student Outcome')
plt.ylabel('Mother\'s Qualification')
plt.show()

# Display the cross-tabulation
print(cross_tab)
```
