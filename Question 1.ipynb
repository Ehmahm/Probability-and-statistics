A. DISTRIBUTION OF CATEGORICAL VARIABLE MEAL TYPES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm


df = pd.read_csv('Q1_hotel_bookings.csv')

# Creating larger figure size
plt.figure(figsize=(10, 6))



# bar plot
sns.countplot(data=df, x='meal', order=df['meal'].value_counts().index)
plt.title('Distribution of Meal Types')
plt.xlabel('Meal Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Statistics
meal_stats = df['meal'].value_counts()
meal_percentages = df['meal'].value_counts(normalize=True) * 100

print("\
Meal Type Distribution:")
print(meal_stats)
print("\
Percentages:")
for meal, percentage in meal_percentages.items():
    print(f"{meal}: {percentage:.2f}%")
```

The most common meal type is BB which accounts for 77.32% of the bookings.

HB  and  Self Catering follow with 12.11% and 8.92%, respectively.

Undefined and Full Board  are the least common, with less than 1% each.


B. DISTRIBUTION OF QUANTITATIVE VARIABLE LEAD TIME

```python
#Distribution of the lead_time variable
plt.figure(figsize=(10, 6))
sns.histplot(df['lead_time'], bins=30, kde=True, color='blue')
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

#Statistics for lead_time
lead_time_stats = df['lead_time'].describe()
print("\
Lead Time Statistics:")
print(lead_time_stats)
```

The average lead time is approximately 104 days, with a standard deviation of 106.86 days.

The median lead time is 69 days, indicating that half of the bookings are made within this time frame.

The lead time ranges from 0 days to a maximum of 737 days.

C. PROBABILITY OF CITY HOTEL BEING BOOKED BY CLIENT FROM COUNTRY ITA IN ARRIVAL MONTH OF FEBRUARY.

```python
# Filtering data for city hotels, Italian clients, and February arrivals
total_city_feb = df[(df['hotel'] == 'City Hotel') & (df['arrival_date_month'] == 'February')].shape[0]
ita_city_feb = df[(df['hotel'] == 'City Hotel') & 
                  (df['country'] == 'ITA') & 
                  (df['arrival_date_month'] == 'February')].shape[0]

# Calculating probability
probability = (ita_city_feb / total_city_feb) * 100

print(f"Total City Hotel bookings in February: {total_city_feb}")
print(f"Italian bookings in City Hotel in February: {ita_city_feb}")
print(f"Probability: {probability:.2f}%")
```
