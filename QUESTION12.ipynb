A)CENTRAL TENDENCY AND DISTRIBUTION OF ANY TWO CONTINUOS VARIABLES
"BILLING AMOUNT" AND "AGE"

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/EHMAH/Documents/healthcare_dataset.csv")

#Central Tendency and Distribution for Billing Amount
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Billing Amount', kde=True)
plt.title('Distribution of Billing Amount')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='Billing Amount')
plt.title('Boxplot of Billing Amount')
plt.tight_layout()
plt.show()

# Statistical measures
billing_stats = df['Billing Amount'].describe()
print("\
Statistical Measures for Billing Amount:")
print(billing_stats)

#Central Tendency and Distribution For Age
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribution of Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='Age')
plt.title('Boxplot of Age')
plt.tight_layout()
plt.show()

# Statistical measures
age_stats = df['Age'].describe()
print("\
Statistical Measures for Age:")
print(age_stats)
```

B)MARGINAL PROBABILITY OF ANY TWO CATEGORICAL VARIABLES.

```python
#contingency table
contingency = pd.crosstab(df['Medical Condition'], df['Gender'], margins=True)
print("\
Contingency Table:")
print(contingency)

#Marginal probabilities
total = len(df)
medical_condition_prob = df['Medical Condition'].value_counts() / total
gender_prob = df['Gender'].value_counts() / total

print("\
Marginal Probabilities for Medical Condition:")
print(medical_condition_prob)
print("\
Marginal Probabilities for Gender:")
print(gender_prob)
```

C)RELATIONSHIP BETWEEN THE MEDICAL CONDITION AND ADMISSION TYPE.

```python
#Contingency table for Medical Condition and Admission Type
contingency_table = pd.crosstab(df['Medical Condition'], df['Admission Type'])

#heatmap to visualize the relationship
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Medical Condition vs Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Medical Condition')
plt.show()

#contingency table
print("Contingency Table for Medical Condition and Admission Type:")
print(contingency_table)
```
